```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

link_V6='https://docs.google.com/spreadsheets/d/e/2PACX-1vRJdgkqLqyXxoY1pYvjZfgCsKd5vLK5gQMbL6L2_mknpumbK3AUt4mZ8hRtCbDa0w/pub?output=xlsx'
circuitos=pd.read_excel(link_V6)
circuitos.set_index(['Id'],inplace=True)

# Vamos a quitar los municipios el agrupamientopoa las 10 zonas Rural o Metropolitana nos daría mejoer

circuitos.describe().T
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

X= circuitos.loc[:, ~circuitos.columns.isin( ['Substation Name','Transformer','MV Feeder Code','Feeder Name MV','REPORTETELCO AnexohojaCondiciones'])]
X2= X.drop(['ZONANUEVA'], axis=1)
#Se trabajará con sólo el SAIDI y SAIFI propio, entonces eliminamos el SAIDI total
X2= X2.loc[:, ~X2.columns.isin( ['SAIDI_EXTERNAL', 'SAIDI_OTHER_CAUSES', 'SAIDI_PLANNED', 'SAIFI_EXTERNAL', 'SAIFI_OTHER_CAUSES', 'SAIFI_PLANNED'])]

#Quitamos la potencia TIEPI
X2= X2.loc[:, ~X2.columns.isin( ['Potencia TIEPI'])]
#Filtramos por cada unidad territorial
X2 = pd.get_dummies(X2)
from sklearn.preprocessing import StandardScaler
scaler = MinMaxScaler() # de 
X2_escalado = scaler.fit_transform(X2)
escalado_df = pd.DataFrame(X2_escalado,columns=X2.columns)

g= sns.clustermap(escalado_df,col_cluster=False,figsize=(20,15),dendrogram_ratio=0.4,cmap='viridis')
plt.show()
plt.savefig('Cluster Jerarquico_ClusterMap_Filas.png', format='png', dpi=500)

X2.describe().T
```

```python
from scipy.cluster.hierarchy import ward, fcluster
from scipy.spatial.distance import pdist
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

dist = pdist(X2_escalado, metric='euclidean')
Z = linkage(dist, method='ward')

distancia_corte=2.5 #con esta distancia se seleccionan la cantidad de clusters
fig, ax3 = plt.subplots(figsize=(30, 8))
d = dendrogram(Z, show_leaf_counts=True, leaf_font_size=9, ax=ax3)
ax3.set_xlabel('Circuitos', fontsize=14)

clusters= fcluster(Z, t=distancia_corte, criterion='distance')
ax3.set_ylabel('Distancia', fontsize=14)
ax3.axhline(y=distancia_corte, c = 'black', linestyle='--', label='altura corte')
plt.show()
print(" ")
print(f'Con una distancia de {distancia_corte} se retienen {max(clusters) } clusters')
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')

cluster_sk = AgglomerativeClustering(n_clusters=12, affinity='euclidean', linkage='complete')
cluster_sk_labels =cluster_sk.fit_predict(X2_escalado)
df_clusterizado=escalado_df.copy()
df_clusterizado['cluster']= cluster_sk_labels

df_clusterizado.head()
```

```python

import math

X_clusterizado= df_clusterizado.copy()
cols_SAIDI= ['SAIDI_EXTERNAL_PROPIO', 'SAIDI_OTHER_CAUSES_PROPIO']
porc_circuitos=0.4

X_clusterizado['suma_saidi']= X_clusterizado[cols_SAIDI].sum(axis=1)
agrupamiento = X_clusterizado.groupby('cluster')
porc_circuitos = 0.4
for cluster, circuitos in agrupamiento:
  print(cluster)
  num_circuitos_cluster= circuitos.shape[0]
  circuitos_a_seleccionar=circuitos.head(math.ceil(num_circuitos_cluster * porc_circuitos))
  circuitos_a_seleccionar.to_excel(f'Circuitos Cluster_{cluster}.xlsx')
```
