```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

circuitos=pd.read_excel('dataset_pfinal_ANS_V2.xlsx')
circuitos.set_index(['Id'],inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

X= circuitos.select_dtypes(include="number")
scaler = StandardScaler()
X_escalado= scaler.fit_transform(X)


from sklearn.decomposition import PCA
pca_model = PCA(n_components=2)
pca_circuitos = pca_model.fit_transform(X_escalado)
varianza_explicada = []
threshold=0.95
rango=range(1,X.shape[1])
for n in rango :
    pca = PCA(n_components=n)
    pca.fit(X_escalado)
    varianza_explicada.append(np.sum(pca.explained_variance_ratio_))
plt.plot(rango,varianza_explicada)
plt.xlabel("Numero de Componentes")
plt.ylabel("varianza_explicada");
plt.axhline(y=threshold,color ="black",linestyle ="--")
varianza_explicada_array= np.array(varianza_explicada)
plt.axvline(x=np.argwhere(varianza_explicada_array > threshold)[0],color ="black",linestyle ="--")
# Numero de compoenentes que explican el 95% de ;a varianza explicada es 64 . Relamente creo que no hay que hacer PCA.

plt.savefig('PCA.png', format='png', dpi=300)
```
